# Segmenting and Clustering Neighborhoods in Toronto

## Part 1

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#Scrapping the Wikipedia page for the table
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
```

```python
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
#Converting the html table to DataFrame
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
# Delete the rows where Borough is 'Not assigned'
df = df[df.Borough != 'Not assigned']
df
```

```python
# Combine the neighbourhoods with equal Postalcode
df = df.groupby(['Postal code','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)
df
```

```python
# Replacing the neighbourhoods which are 'Not assigned'
df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned',df['Borough'], df['Neighborhood'])
df
```

```python
# Shape the dataframe
df.shape
```

## Part 2

```python
#read the data
coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
coordinates.head()
```

```python
#Define columns at coordinates
coordinates.columns = ['Postal Code','Latitude','Longitude']
coordinates.head()
```

```python
#Define columns at df
df.columns = ['Postal Code','Borough','Neighborhood']
df.head()
```

```python
#Merge the Datasets coordinates and df
df = pd.merge(df, coordinates, on = 'Postal Code')
df.head()
```

## Part 3

```python
#Getting all the rows from the data frame which contains Toronto in their Borough.
df2 = df[df['Borough'].str.contains('Toronto',regex=False)]
df2
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library
```

```python
#Visualisation
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df2['Latitude'],df2['Longitude'],df2['Borough'],df2['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
#kmeans clustering
from sklearn.cluster import KMeans
```

```python
k=5
toronto_clustering = df2.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df2.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# create map with final cluster
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df2['Latitude'], df2['Longitude'], df2['Neighborhood'], df2['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
